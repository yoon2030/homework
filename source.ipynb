```python
import pandas as pd
import requests 
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
url = "http://www.ekapepia.com/priceStat/period/periodPeriodPrice.do?menuId=menu100009&boardInfoNo="
req = requests.get(url, verify=False)
res=req.text
soup = BeautifulSoup(res,'html.parser')
keywords = soup.find_all('div', class_='table-wrap')
keywords = [line.get_text().strip() for line in keywords]
str3=keywords[0]
str3 = str3.split('\n')
list0=str3[57::18]
list1=str3[58::18]
list2=str3[62::18]
list3=str3[63::18]
list4=str3[64::18]
list5=str3[65::18]
list1[19]='0'
list2[19]='0'
for i in range(len(list1)) :
    list1[i] = int(list1[i].replace(',',''))
    list2[i] = int(list2[i].replace(',',''))
    list3[i] = int(list3[i].replace(',',''))
    list4[i] = int(list4[i].replace(',',''))
    list5[i] = int(list5[i].replace(',',''))
s = pd.DataFrame({
    '1++등급':list1,
    '1+등급':list2,
    '1등급':list3,
    '2등급':list4,
    '3등급':list5,
    'day':list0,
})
df= s.groupby('day')['1++등급','1+등급','1등급','2등급','3등급'].describe()
print(df)
df['1++등급'].plot()
df['1+등급'].plot()
df['1등급'].plot()
df['2등급'].plot()
df['3등급'].plot()
```
